```python
prices = [100,200,300,400,500,600]
```

```python
for p in prices:
  print(p,end=' ')
```

```python
total = 0
for x in prices:
  total = total + x
print(total)
```

```python
money = 1000
target = 2000
years = 0

while money < target:
  money = money + (money * 0.08)
  years = years + 1
print(years)
```

```python
boosted = [round(p*1.10) for p in prices]
print(boosted)
```

```python
salaries = [5500, 6200, 6800, 7100, 4900]
def has_high_salary(salaries):
  for x in salaries:
    if x >= 600:
      print('above target')
```

```python
has_high_salary(salaries)
```

```python
#use for loop to find how many above 100
#for loop use panni traverse pannalam so can chck along th way and if higher

def count_of_higher(salaries):
  count = 0
  for x in salaries:
    if x > 100:
      count = count + 1
  return count
```

```python
count_of_higher(salaries)
```

```python
rounded_salaries = [round(p) for p in salaries]
print(rounded_salaries)
```
