```python
prices = [100,200,400,600,700]
```

```python
print('first day:',prices[0])
```

```python
print('last day:',prices[-1])
```

```python
print('mid week:',prices[1:-1])
```

```python
print('how many days:',len(prices))
```

```python
print('highest:',max(prices))
```

```python
print('lowest:',min(prices))
```

```python
print('average:',sum(prices) / len(prices))
```

```python
prices.append(900)
```

```python
print('new list:',sorted(prices))
```
