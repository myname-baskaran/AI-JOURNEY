```python
ticker = '  dbs  '
clean = ticker.strip().upper()
print(clean)
```

```python
price = 35.0
print(f'{clean} is trading at ${price}')
```

```python
note = "Buy DBS, hold OCBC, sell UOB"
print(note.split())
```

```python
stock = {"ticker": "DBS", "price": 35.20, "sector": "Banking"}
print(stock['price'])
stock['price'] = 69.69
stock['dividend']=0.80
print(stock)
```

```python
portfolio = {"DBS": 35.20, "OCBC": 14.80, "UOB": 31.50}
for name in portfolio:
  print(f'{name}:stock price ${portfolio[name]}')
```

```python
me = {"name": "abi", "goal_salary": 6500, "ord": "May 2027"}
print(f'my name is {me['name']} and my salary goal is {me['goal_salary']} and i ord on {me['ord']}')
```

```python
stocks = "AAPL,GOOG,TSLA,NVDA"
print(len(stocks.split(',')))
```

```python
portfolio = {"DBS": 35.20, "OCBC": 14.80, "UOB": 31.50}
for stock in portfolio:
  if portfolio[stock] > 20:
    print(stock)
```
