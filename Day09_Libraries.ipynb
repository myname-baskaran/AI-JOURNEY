```python
import math
print(math.sqrt(144))
print(math.pi)
print(round(math.pi,2))
```

```python
import numpy as np
prices = np.array([100,200,300,400,500])
print('average price:{}'.format(np.mean(prices)))
print('biggest jump:{}'.format(np.max(prices)))
print('standard deviation:{}'.format(round(np.std(prices),2)))
```

```python
import random
for day in range(5):
  move = random.uniform(-3,3)
  print('day{}:{}'.format(day,round(move,2)))
```

```python
import math
print(math.floor(7.8))
print(math.ceil(7.2))
```
