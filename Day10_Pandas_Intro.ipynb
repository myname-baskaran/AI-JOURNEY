```python
import pandas as pd

stocks = pd.DataFrame({"Ticker": ["DBS", "OCBC", "UOB"],
    "Price": [35.20, 14.80, 31.50],
    "Sector": ["Banking", "Banking", "Banking"]
})
```

```python
print(stocks)
```

```python
print(stocks.head())
print(stocks.shape)
print(stocks['Price'])
```

```python
print(stocks['Price'].mean())
print(stocks['Price'].max())
```

```python
portfolio = pd.DataFrame({"Ticker": ["DBS", "OCBC", "UOB"],
    "Price": [35.20, 14.80, 31.50],
    "Sector": ["Banking", "Banking", "Banking"]})

print(portfolio['Price'].mean())
```
