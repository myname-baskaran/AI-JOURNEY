```python
def future_value(principal,interest_rate,years):
  '''returns investment rate after compounding'''
  return principal*(1+interest_rate)**years
```

```python
print(round(future_value(1000,0.05,10)))
```

```python
def monthly_to_annual(monthly,bonus_months = 1):
  return monthly * (12 + bonus_months)
```

```python
print(monthly_to_annual(1000))
print(monthly_to_annual(1000,5))
```

```python
def savings_goal(monthly_save,months):
  return monthly_save * months
```

```python
print(savings_goal(100,12))
```

```python
def take_home(salary, tax_rate=0.15):
  return salary * (1-tax_rate)
```

```python
print(take_home(100000))
print(take_home(100000,0.20))
```
