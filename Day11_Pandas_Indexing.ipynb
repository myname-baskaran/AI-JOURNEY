```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Ticker": ["DBS", "OCBC", "UOB", "SIA", "Grab", "SeaLtd"],
    "Price":  [35.20, 14.80, 31.50, 6.40, 4.10, 78.30],
    "Sector": ["Banking", "Banking", "Banking", "Airline", "Tech", "Tech"]
})
print(df)
```

```python
print(df.iloc[0])
```

```python
print(df.iloc[:3])
```

```python
print(df.iloc[-1])
```

```python
expensive = df.loc[df.Price > 20]
print(expensive)
```

```python
tech = df.loc[df.Sector.isin(['Tech'])]
print(tech)
```

```python
#Use iloc to print the 2nd and 3rd rows of df (remember: counting starts at 0).

second_row = df.iloc[[1,2],:]
print(second_row)
```

```python
#Use conditional selection to show only the stocks in the "Banking" sector
banks = df.loc[df.Sector.isin(['Banking']) & df.Ticker]
print(banks)
```

```python
#Combine two conditions: show stocks that are Tech and priced over 50.
df.loc[(df.Sector.isin(['Tech'])) & (df.Price > 50)]
```
