```python
#A simple expense analyzer that totals spending, finds the average, and flags months over budget — built using Python functions, loops, lists, and conditionals.
```

```python
#monthly expenses for the year
expenses = [1200, 1500, 980, 2100, 1750, 1300, 1600, 1100, 2400, 1900, 1050, 1450]
budget = 1500
```

```python
#total amount spent
def total_spent(expenses_list):
  total = 0
  for x in expenses_list:
    total = total + x
  return total
```

```python
def average_spent(expenses_list):
  return total_spent(expenses_list)/ len(expenses_list)
```

```python
def months_over_budget(expenses_list,budget):
  count = 0
  for y in expenses_list:
    if y > budget:
      count+=1
  return count
```

```python
print('total spent:',total_spent(expenses))
print('average spent:',average_spent(expenses))
print('months over budget:',months_over_budget(expenses,budget))
```

```python
def over_budget_months(expenses_list,budget):
  overbudget_list = []
  for n in expenses_list:
    if n > budget:
      overbudget_list.append( n )
  return overbudget_list
```

```python
def full_list(expenses_list,budget):
  #print each month with s no and to also check if the budget is also high or low
  for k in range(len(expenses_list)):
    if expenses_list[k] > budget:
      print('month',k+1,':',expenses_list[k],'-','over')
    else:
      print('month',k+1,':',expenses_list[k],'-','ok')
```

```python
print(full_list(expenses,budget))
```
