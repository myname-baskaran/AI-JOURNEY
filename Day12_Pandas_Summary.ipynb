```python
import pandas as pd

df = pd.DataFrame({
    "Ticker": ["DBS", "OCBC", "UOB", "SIA", "Grab", "SeaLtd"],
    "Price":  [35.20, 14.80, 31.50, 6.40, 4.10, 78.30],
    "Sector": ["Banking", "Banking", "Banking", "Airline", "Tech", "Tech"]
})
```

```python
df.Price.describe()
```

```python
print(df.Price.mean())
```

```python
print(df.Ticker.unique())
```

```python
print(df.Ticker.value_counts())
```

```python
df['price_usd'] = df.Price.map(lambda p: round(p * 0.74,2))
```

```python
df["Price"].describe()
```

```python
df.Sector.value_counts()
```

```python
df['Price_rounded'] = df.Price.map(lambda p: round(p))
```
