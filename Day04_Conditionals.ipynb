```python
def loan_approval(monthly_income,credit_score):
  if credit_score > 500 and monthly_income > 1000:
    print('approved')
  elif credit_score > 500:
    print('review needed')
  else:
    print('denied')
```

```python
loan_approval(1000,500)
```

```python
salary = 6500
print(salary >= 6500)
print(salary > 6000 and salary < 7000)
```

```python
def tax_bracket(income):
  if income > 8000:
    print('high')
  elif income > 4000:
    print('medium')
  else:
    print('low')
```

```python
tax_bracket(100)
tax_bracket(1000000)
```

```python
def can_invest(savings,emergency_funds):
  if savings > 5000 and emergency_funds :
    print('go ahead')
  else:
    print('wait')
```

```python
can_invest(10000,True)
```
